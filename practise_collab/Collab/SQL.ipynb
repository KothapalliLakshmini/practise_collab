```python
import sqlite3



books_data = [
    ('Python Programming', 'John Smith', 599.99, 25),
    ('Data Science Handbook', 'Jane Doe', 899.50, 15),
    ('Machine Learning Basics', 'Alan Turing', 1299.00, 10),
    ('SQL Essentials', 'Edgar Codd', 499.99, 30),
    ('Web Development', 'Tim Berners', 799.00, 20)
]

conn = sqlite3.connect('bookstore.db')
cursor = conn.cursor()
cursor.execute(''' CREATE TABLE IF NOT EXISTS Books (
book_id INTEGER Primary Key AUTOINCREMENT,
title   TEXT Not Null,
author  TEXT Not Null,
price   REAL Not Null,
stock_quantity INTEGER Default 0
)
''')

conn.commit()

cursor.executemany('''
INSERT INTO Books (title, author, price, stock_quantity)
VALUES (?, ?, ?, ?)
''', books_data)

conn.commit()
conn.close()
```

```python
customers_data = [
    ('Rahul Sharma', 'rahul@email.com', 'Mumbai', '2024-01-15'),
    ('Priya Patel', 'priya@email.com', 'Delhi', '2024-01-20'),
    ('Amit Kumar', 'amit@email.com', 'Bangalore', '2024-02-01'),
    ('Sneha Reddy', 'sneha@email.com', 'Hyderabad', '2024-02-10'),
    ('Vikram Singh', 'vikram@email.com', 'Mumbai', '2024-02-15')
]

conn = sqlite3.connect('bookstore.db')
cursor = conn.cursor()
cursor.execute(''' CREATE TABLE IF NOT EXISTS Customers  (
customer_id INTEGER Primary Key AUTOINCREMENT,
name TEXT Not Null,
email TEXT Unique Not Null,
city TEXT,
join_date TEXT
)
''')

conn.commit()

cursor.executemany(''' INSERT INTO Customers('name', 'email', 'city', 'join_date')
Values(?,?,?,?)''',customers_data)

conn.commit()

conn.close()
```

```python

orders_data = [
    (1, 1, 2, '2024-03-01', 1199.00),
    (1, 2, 1, '2024-03-02', 899.50),
    (2, 1, 1, '2024-03-03', 599.99),
    (2, 3, 1, '2024-03-05', 1299.00),
    (3, 4, 3, '2024-03-07', 1499.97),
    (4, 2, 1, '2024-03-10', 899.50),
    (5, 5, 2, '2024-03-12', 1598.00)
]

conn = sqlite3.connect('bookstore.db')

cursor = conn.cursor()

cursor.execute('''
CREATE TABLE IF NOT EXISTS Orders (
    order_id INTEGER PRIMARY KEY AUTOINCREMENT,
    customer_id INTEGER,
    book_id INTEGER,
    quantity INTEGER NOT NULL,
    order_date TEXT NOT NULL,
    total_amount REAL,
    FOREIGN KEY (customer_id) REFERENCES Customers(customer_id),
    FOREIGN KEY (book_id) REFERENCES Books(book_id)
)
''')

conn.commit()

cursor.executemany(''' INSERT INTO Orders('customer_id', 'book_id', 'quantity', 'order_date', 'total_amount')
Values(?,?,?,?,?)''',orders_data)
conn.commit()


conn.close()
```

```python
conn = sqlite3.connect('bookstore.db')

df = pd.read_sql("SELECT customer_id,COUNT(*) FROM Orders Group By customer_id Order By total_amount", conn)
print(df.head())
conn.close()
```

```python
conn = sqlite3.connect('bookstore.db')
df = pd.read_sql("""
SELECT c.name, COUNT(o.order_id) AS total_orders
FROM Customers c
JOIN Orders o ON c.customer_id = o.customer_id
GROUP BY c.customer_id, c.name
ORDER BY total_orders DESC
""", conn)

print(df.head())
conn.close()
```

```python
conn = sqlite3.connect('bookstore.db')

books_df = pd.read_sql("SELECT * FROM Books", conn)
orders_df = pd.read_sql("SELECT * FROM Orders", conn)
customers_df = pd.read_sql("SELECT * FROM Customers", conn)

print(books_df.head(3))
print(orders_df.head(3))
print(customers_df.head(3))

merge_df = orders_df.merge(customers_df, on='customer_id').merge(books_df, on='book_id')
merge_df[[
    'order_id',
    'name',
    'city',
    'title',
    'quantity',
    'total_amount'
]]

print(merge_df.head())

avg_order_value = orders_df['total_amount'].mean()
print("Average Order Value:", avg_order_value)

total_orders = merge_df.groupby('city')['quantity'].count()
print("Total Orders per City:")
print(total_orders)

most_ordered = merge_df.groupby('title')['quantity'].sum().idxmax()
print("Most Ordered Book:", most_ordered)
conn.close()
```
