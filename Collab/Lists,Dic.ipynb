```python
import numpy as np


data_array = np.random.seed(42)
scores = np.random.randint(50,101, size =(5,4))

print(f"Scores : {scores}\n")
print(f"Scores of 3rd student and 2nd subject {scores[2,1]}\n")
print(f"Scores of last 2 students {scores[-2:,:]}\n")
print(f"Scores of first 3 students in subjects 2 and 3: {scores[0:3,1:3]}\n")

average_score = np.round(scores.mean(axis = 0),2)

curve = np.array([5,3,7,2])
curved_scores = scores + curve
curved_scores = np.clip(curved_scores,0,100)

print(f"curved_scores :{curved_scores}")

row_max = curved_scores.max(axis  = 1)
print("Row-wise max:", row_max)

row_min = curved_scores.min(axis=1, keepdims=True)
row_max = curved_scores.max(axis=1, keepdims=True)

normalized = (curved_scores - row_min)/(row_max - row_min)

print("Normalized Scores:\n", normalized)

max_index = np.unravel_index(np.argmax(normalized), normalized.shape)

print("Highest value at (student, subject):", max_index)

high_scores = curved_scores[curved_scores > 90]

print("Scores > 90:", high_scores)

```

```python
shopping_list = [
    {"item": "Milk", "price": 50},
    {"item": "Bread", "price": 30},
    {"item": "Eggs", "price": 60},
    {"item": "Rice", "price": 120}
]

# Task 1: Add and Remove Items

# Add new item
shopping_list.append({"item": "Butter", "price": 80})

# Remove first item
shopping_list.pop(0)

# Print total items
print(f"Total items: {len(shopping_list)}")
print()

# Task 2: Calculate Total Cost

# Calculate total cost
total_cost = 0
for item in shopping_list:
    total_cost += item["price"]

# Find most expensive item
max_price = 0
expensive_item = ""

for item in shopping_list:
    if item["price"] > max_price:
        max_price = item["price"]
        expensive_item = item["item"]

print(f"Most expensive: {expensive_item} - Rs. {max_price}")
print(f"Total cost: Rs. {total_cost}")
print()

# Task 3: Create Summary Dictionary

# Calculate average
average_price = round(total_cost / len(shopping_list), 2)

# Create summary dictionary
summary = {
    "total_items": len(shopping_list),
    "total_cost": total_cost,
    "average_price": average_price
}

print(f"Summary: {summary}")
```

```python
# STUDENT GRADE BOOK
# Using Lists, Dictionaries, Tuples, and Sets

# List - student names (ordered, mutable)
students = ["Raj", "Priya", "Amit", "Sneha", "Karan"]

# Tuple - subject names (immutable, won't change)
subjects = ("Math", "Science", "English")

# Dictionary - store marks (key-value pairs for easy access)
marks = {
    "Raj": {"Math": 85, "Science": 78, "English": 90},
    "Priya": {"Math": 92, "Science": 88, "English": 85},
    "Amit": {"Math": 70, "Science": 75, "English": 68},
    "Sneha": {"Math": 95, "Science": 90, "English": 92},
    "Karan": {"Math": 80, "Science": 82, "English": 78}
}

# Set - unique grades (no duplicates)
grades = set()

# Display all students
print(f"All Students: {students}")

# Slicing - first 3 students
print(f"First 3 Students: {students[0:3]}\n")

# Calculate averages and assign grades
top_student = ""
highest_avg = 0

for student in students:
    # Get total marks
    total = marks[student]["Math"] + marks[student]["Science"] + marks[student]["English"]

    # Calculate average
    average = total / 3

    # Assign grade
    if average >= 85:
        grade = "A"
    elif average >= 70:
        grade = "B"
    else:
        grade = "C"

    # Add to set
    grades.add(grade)

    # Print student info
    print(f"{student} - Average: {average:.2f} - Grade: {grade}")

    # Track top student
    if average > highest_avg:
        highest_avg = average
        top_student = student

# Display results
print(f"\nTop Student: {top_student}")
print(f"Unique Grades: {grades}")
```

```python
# Step 1: Initialize variables
total_seats = 350
remaining_seats = 350
total_bookings = 0
tickets_sold = 0
rejected_bookings = 0

# Step 2: Start main booking loop
while True:
    tickets = int(input("Enter number of tickets (0 to exit): "))

    # Step 3: Exit condition
    if tickets == 0:
        break

    # Step 4: Validate ticket count
    if tickets < 1 or tickets > 15:
        print("Invalid number of tickets. Must be between 1 and 15.")
        continue  # skip this iteration

    # Step 5: Check seat availability
    if tickets > remaining_seats:
        print("BOOKING REJECTED - Not enough seats")
        rejected_bookings += 1
        continue

    # Step 6: Collect ages
    ages = []
    for i in range(tickets):
        age = int(input(f"Enter age of person {i+1}: "))
        ages.append(age)

    # Step 7: Validate age restriction
    invalid_age = False
    for age in ages:
        if age < 12:
            invalid_age = True
            break

    # Step 8: Reject if any age < 12
    if invalid_age:
        print("BOOKING REJECTED - Age restriction")
        rejected_bookings += 1
        continue

    # Step 9: Confirm booking
    print(f"BOOKING CONFIRMED - {tickets} tickets")

    total_bookings += 1
    tickets_sold += tickets
    remaining_seats -= tickets

    # Step 10: Stop if theatre is full
    if remaining_seats == 0:
        print("Theatre is now full!")
        break

# Step 11: Final Report
print("\nFINAL REPORT")
print("Total Bookings:", total_bookings)
print("Total Tickets Sold:", tickets_sold)
print("Rejected Bookings:", rejected_bookings)
print("Remaining Seats:", remaining_seats)
```
