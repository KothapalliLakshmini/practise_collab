```python
#what is the change in mean avergae price of honey from 1995 to 2021

y = df.groupby('year')['average_price'].mean().reset_index()
y
```

```python
#year product - highest production
#now let infer which state was having highest contribution in that year

highest_prod_year = df.groupby('year')['production'].sum().idxmax()

int(highest_production_year)
```

```python
high_year = df[df['year'] == highest_production_year]
state  = high_year.groupby('state')['production'].sum().idxmax()
state
```
